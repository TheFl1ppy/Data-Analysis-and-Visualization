### Форматирование -приведение данных к приемлемому формату

```python
#df3.fillna(0, inplace = True)
df3.drop('ndvi_ne',axis = 1, inplace = True )
df3.drop('ndvi_nw',axis = 1, inplace = True )
df3.drop('ndvi_se',axis = 1, inplace = True )
df3.drop('ndvi_sw',axis = 1, inplace = True )
df3.drop('reanalysis_air_temp_k',axis = 1, inplace = True )
df3.drop('reanalysis_avg_temp_k',axis = 1, inplace = True )
df3.drop('reanalysis_dew_point_temp_k',axis = 1, inplace = True )
df3.drop('reanalysis_max_air_temp_k',axis = 1, inplace = True )
df3.drop('reanalysis_min_air_temp_k',axis = 1, inplace = True )

df3.dropna(inplace=True)
```

### были выброшены строки в которых отсутствовало больше половины значений, так же столбцы которые не нужны для обучающей выборки

```python
df3['year'] = pd.to_datetime(df3['year'], format='%Y')
df3['week_start_date'] = pd.to_datetime(df3['week_start_date'], format='%Y-%m-%d')
```
